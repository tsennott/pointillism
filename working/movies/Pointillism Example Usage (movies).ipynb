# Multiprocessing batch

```python
%run pointillism.py
import imageio
from multiprocessing import Pool
import os
```

```python
location = 'movies/batch/redos/'
filenames = []
for file in os.listdir(location):
    if not file.startswith('._'):
        if file.endswith(".mp4") | file.endswith(".mov"):
            if not (os.path.isfile(location + file + ' - out.mp4')
                    | file.endswith('out.mp4')):
                filenames.append(location + file)
```

```python
filenames
```

```python
for file in filenames:
    reader = imageio.get_reader(file)
    fps = reader.get_meta_data()['fps']

    # Load up whole file to check validity, use first frame to make points, then discard
    frames = []
    try:
        for frame in reader:
            frames.append(frame)
    except:
            print('error with file ' + file)
            continue
    image = Image.fromarray(frames[0])
    point=pointillize(image=image, border=0, reduce_factor = 1, increase_factor = 2)
    locations = point._generateRandomPoints(10e4)
    frames = []
    
    writer = imageio.get_writer(file + ' - out.mp4', fps=fps)
    
    reader_chunk_size = 96
    reader_chunks = ([[x, min(x+reader_chunk_size, len(reader))] 
                              for x in range(0,len(reader),reader_chunk_size)])

    start = time.time()
    print(file, end='... ')
    for i, chunk in enumerate(reader_chunks):

        frames = []
        for x in range(chunk[0], chunk[1]):
            frames.append(reader.get_next_data())

        n_processes = 8
        chunk = len(frames)//n_processes
        if chunk == 0: continue
        frame_chunks = [frames[x:x+chunk] for x in range(0,len(frames),chunk)]

        def f(frames):
            im_out = []
            then = time.time()
            for j, frame in enumerate(frames):
                image = Image.fromarray(frame)
                point=pointillize(image=image, border=0, reduce_factor = 1, increase_factor = 2)
                point.plotRecPoints(40, multiplier=1, fill=True)
                point.plotRandomPointsComplexity(n=10e4, constant=0.004,
                                                 power=2, locations=locations)
                im_out.append(np.array(point.out))
            return im_out

        p = Pool(n_processes)

        out = p.map(f,frame_chunks)

        out_frames = []
        for frames in out:
            for frame in frames:
                out_frames.append(frame)

        for frame in out_frames:
            writer.append_data(frame[:,:,:])  

        print(i, end=' ')
        p.close()
        p.join()
        
    writer.close()
    end = time.time()
    print('took %0.2f minutes' % ((end-start)/60))
```
